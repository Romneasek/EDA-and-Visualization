```python
# Import the required Libraries
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create data

def create_sample_chart():
    x_series = np.arange(10)
    y_series = x_series ** 2

    #Plotting to our canvas in memory
    plt.plot(x_series, y_series)
    #Title of our canvas
    plt.title('String Title Here')
    #X-axis label
    plt.xlabel('X Axis Title Here')
    #Y_axis label
    plt.ylabel('Y Axis Title Here')

    #Showing what we plotted
    plt.show()
```

```python
create_sample_chart()
```

```python
def sine_wave():
  x = np.arange(0,10,0.1)
  y = np.sin(x)
   #Plotting to our canvas in memory
  plt.plot(x,y)
    #Title of our canvas
  plt.title('y = sin(x)')
    #X-axis label
  plt.xlabel('X Value')
    #Y_axis label
  plt.ylabel('Y Value')

    #Showing what we plotted
  plt.show()
```

```python
sine_wave()
```

```python
weather_df = pd.read_csv('/content/weather_2012.csv', parse_dates=True, index_col='Date/Time')
weather_df.head(5)
```

```python
import matplotlib.pyplot as plt
import calendar

def bar_plot():
    weekly_data = weather_df.groupby(weather_df.index.dayofweek).mean()
    
    plt.bar(weekly_data.index, weekly_data['Visibility (km)'])

    plt.title('Visibility by week, 2012')
    plt.xlabel('Day of week')
    plt.ylabel('Visibility (km)')

    plt.xticks(weekly_data.index, calendar.day_abbr, rotation=45)

    plt.show()
```

```python
bar_plot()
```

```python
def two_plots():
    x = np.array([0, 1, 2, 3, 4, 5])
    y = x ** 2

    # Create Figure (empty canvas)
    fig = plt.figure()

    # Add set of axes to figure
    axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
    axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes 
    # (0.2x left, 0.5x bottom) to (0.4x width, 0.3x height)

    # Larger Figure Axes 1
    axes1.plot(x, y, 'b')
    axes1.set_xlabel('X_label_axes1')
    axes1.set_ylabel('Y_label_axes1')
    axes1.set_title('Axes 1 Title')

    # Insert Figure Axes 2
    axes2.plot(y, x, 'r')
    axes2.set_xlabel('X_label_axes2')
    axes2.set_ylabel('Y_label_axes2')
    axes2.set_title('Axes 2 Title');    
```

```python
two_plots()
```

```python
def sub_plots_example():
  # plot 1
  x = np.arange(0,10)
  y = x **2
  plt.subplot(2,2,1)
  plt.plot(x,y, c='r')
  
  # plot 2
  x= np.array([0,1,2,4])
  y = np.array([10,12,13,14])
  plt.subplot(2,2,2)
  plt.plot(x,y)

  #plot 3
  x = np.arange(0,4)
  y = np.arange(2,6)
  plt.subplot(2,2,3)
  plt.plot(x,y)

  #plot 4

  x = np.arange(0,10)
  y = x **2
  plt.subplot(2,2,4)
  plt.plot(x,y, 'g--')


  plt.show()

sub_plots_example()
    
```
